```python
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd
import regex as re
```

```python
 # Собираем все страницы со списком авторов

 all_pages = ['https://www.aozora.gr.jp/index_pages/person_a.html',
              'https://www.aozora.gr.jp/index_pages/person_ka.html',
              'https://www.aozora.gr.jp/index_pages/person_sa.html',
              'https://www.aozora.gr.jp/index_pages/person_ta.html',
              'https://www.aozora.gr.jp/index_pages/person_na.html',
              'https://www.aozora.gr.jp/index_pages/person_ha.html',
              'https://www.aozora.gr.jp/index_pages/person_ma.html',
              'https://www.aozora.gr.jp/index_pages/person_ya.html',
              'https://www.aozora.gr.jp/index_pages/person_ra.html',
              'https://www.aozora.gr.jp/index_pages/person_wa.html']
```

Задача - собрать ссылки на всех писателей из списка А.

```python
url_authors = 'https://www.aozora.gr.jp/index_pages/person_a.html'
```

```python
page_a = rq.get(url_authors)
print(page_a)
```

```python
page_a.encoding = 'utf-8'
print(page_a.text)
```

```python
soup_a = BeautifulSoup(page_a.text, features="html.parser")
soup_a
```

```python
#Надо исключить писателей, у которых есть тэг strong , потому что их произведения открыто не распространяются
link_author = []
for i in soup_a.find_all('a'):
  if i.parent.name == 'li' and i.get('strong') is None:
    link_author.append(i.get('href'))

link_author
```

```python
main_part_author = 'https://www.aozora.gr.jp/index_pages/'
```

```python
# Собираем из двух частей ссылку на каждого автора из списка А
full_links_author = []

for link in link_author:
  link = main_part_author + link
  full_links_author.append(link)

full_links_author
```

Задача - собрать ссылки на всех авторов в библиотеке

```python
all_authors = []
for page in all_pages:
  url = all_pages[0]
  page0 = rq.get(url)
  page0.encoding = 'utf-8'
  soup0 = BeautifulSoup(page0.text, features="html.parser")

  link_author = []
  for i in soup0.find_all('a'):
    if i.parent.name == 'li' and i.get('strong') is None:
      link_author.append(i.get('href'))

  main_part_author = 'https://www.aozora.gr.jp/index_pages/'

  full_links_author = []
  for link in link_author:
    link = main_part_author + link
    full_links_author.append(link)

  all_authors.append(full_links_author)
```

```python
all_authors_list = []
for sublist in all_authors:
    all_authors_list.extend(sublist)
```

Задача - собрать ссылки на все карточки всех писателей.

```python
all_cards = []
for author in all_authors_list:
  url = all_authors_list[0]
  page = rq.get(url)
  page.encoding = 'utf-8'
  soup = BeautifulSoup(page.text, features="html.parser")

  part_link = []
  for i in soup.find_all('a'):
    if i.parent.name == 'li' and 'card' in i.get('href'):
      part_link.append(i.get('href'))

  corrected_part_link = [i.replace('..','') for i in part_link]

  main_part = 'https://www.aozora.gr.jp'

  full_links = []
  for link in corrected_part_link:
    link = main_part + link
    full_links.append(link)

  all_cards.append(full_links)
```

```python
all_cards_list = []
for sublist in all_cards:
    all_cards_list.extend(sublist)
```

Задача - собрать ссылки на все произведения всех авторов.

```python
all_works = []
for card in all_cards_list:
  url2 = all_cards_list[0]
  page2 = rq.get(url2)
  page2.encoding = 'utf-8'
  soup2 = BeautifulSoup(page2.text, features="html.parser")

  final_link = []
  for i in soup2.find_all('a'):
    if i.parent.name == 'td' and 'files' in i.get('href') and 'html' in i.get('href'):
      final_link.append(i.get('href'))

  end_link = re.sub(r'^\.', '', final_link[0])

  link0 =re.sub(r'\/card\d*\.html', '', all_cards_list[0])+end_link

  all_works.append(link0)
```

Задача - написать функцию, которая соберет информацию по всем произведениям всех авторов в библиотеке.

```python
def parsed_data (url0):
  page0 = rq.get(url0)
  page3.encoding = 'Shift_JIS'
  soup0 = BeautifulSoup(page0.text, features="html.parser")
  title = soup0.find_all('h1')[0].text
  author = soup0.find_all('h2')[0].text
  text = soup3.find_all('body')[0].text
  #final_text = ' '.join(text)
  #final_text = final_text.replace('\xa0', ' ')
  return url0, author, title, text
```

```python
corpus = [] # список с новостями

for work in all_works:
  processed = parsed_data(work)
  corpus.append(processed)
```
